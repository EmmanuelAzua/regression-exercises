```python
def wrangle_telco():
    '''This function imports the necessary functions to turn the telco_churn database from SQL into a Pandas DataFrame.
    In addition, it cleans it to include only the customer_id, monthly_charges, tenure, and total_charges variables
    for all customers with a two year contract.'''
    #First, we import the credentials required to access our target database:
    from env import username, password, host
    # Second, we import the pre-defined function to display the telco_churn database tables we need and turns it into
    # a Pandas Dataframe to facilitate preparation:
    from wrangle import telco_churn_data
    # Third, we turn the newly imported and transformed DataFrame into a local CSV file, which will speed the process
    # of import and cleaning:
    from wrangle import churn_data_csv
    # Fourth, we define the DataFrame name, which will help us manipulate and process the information into the formats
    # we need:
    telco_df = churn_data_csv()
    # Sixth, we clean and reorganize the DataFrame to contain only the variables we need, in the order we need them:
    telco_df = telco_df[['customer_id', 'monthly_charges', 'tenure', 'total_charges', 'contract_type']]
    # Seventh, we filter our observations to only include those that have a two-year contract in our DataFrame:
    telco_df = telco_df[telco_df.contract_type == "Two year"]
    # Eigth, we drop the contract_type variable to ensure our DataFrame only has the necessary variables:
    telco_df = telco_df[['customer_id', 'monthly_charges', 'tenure', 'total_charges']]
    return telco_df
```
